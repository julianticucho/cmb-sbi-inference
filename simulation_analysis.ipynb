```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns
import camb
from camb import model, initialpower
import scienceplots
import pandas as pd

plt.style.use(['science', 'bright'])
plt.rcParams['figure.dpi'] = 300
```

```python
# Cargar las simulaciones y los parámetros cosmológicos
Cl_sims = torch.load("data/raw/cmb_angular_sims.pt")
params = torch.load("data/raw/angular_sim_params.pt")

Cl_sims = Cl_sims.numpy()
params = params.numpy()

print(Cl_sims.shape)
print(params.shape)
```

```python
mean_Cl = np.mean(Cl_sims, axis=0)
std_Cl = np.std(Cl_sims, axis=0)

plt.figure(figsize=(12, 8))
plt.fill_between(range(len(mean_Cl)), mean_Cl[:, 0] - std_Cl[:, 0], mean_Cl[:, 0] + std_Cl[:, 0], alpha=0.3, label='Desviación Estándar')
for i in range(10):  
    plt.plot(Cl_sims[i, :, 0], label=f'Realización {i+1}')
plt.xlabel('Multipolo $\ell$')
plt.ylabel('$C_\ell^{TT}$ [$\mu K^2$]')
plt.title('Espectros de Potencia Angular $C_\ell^{TT}$')
plt.legend()
plt.show()
```
