```python
import torch
import os
import matplotlib.pyplot as plt
import scienceplots
import numpy as np

plt.style.use(['science', 'bright'])
plt.rcParams['figure.dpi'] = 300
```

```python
from src.config import PATHS

simulations = torch.load(os.path.join("..",PATHS["simulations"], "Cls_TT_noise_25000.pt"), weights_only=True)
theta, Cls = simulations["theta"], simulations["x"]

mean_Cl = np.mean(Cls.numpy(), axis=0)
std_Cl = np.std(Cls.numpy(), axis=0)

plt.figure(figsize=(12, 8))
plt.fill_between(range(len(mean_Cl)), mean_Cl - std_Cl, mean_Cl + std_Cl, alpha=0.3, label='std')
for i in range(3):  
    plt.plot(Cls[i, :], label=f'Cosmología {i+1}')
plt.xlabel('Multipolo $\ell$')
plt.ylabel('$C_\ell^{TT}$ [$\mu K^2$]')
plt.legend()
plt.show()
```
