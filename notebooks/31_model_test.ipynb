```python
import matplotlib.pyplot as plt
import numpy as np
import os
import torch
from pathlib import Path

def find_repo_root(start: Path = Path.cwd()) -> Path:
    for parent in [start] + list(start.parents):
        if (parent / ".git").exists():
            return parent
    return start  

repo_root = find_repo_root()
os.chdir(repo_root)
```
