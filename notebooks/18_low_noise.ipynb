```python
import torch
import os
import matplotlib.pyplot as plt
import scienceplots
import numpy as np

plt.style.use(['science', 'bright'])
plt.rcParams['figure.dpi'] = 300
```

```python
from src.config import PATHS

simulations1 = torch.load(os.path.join("..",PATHS["simulations"], "Cls_TT_reduced_prior_low_noise_50000.pt"), weights_only=True)
simulations2 = torch.load(os.path.join("..",PATHS["simulations"], "Cls_TT_reduced_prior_noise_50000.pt"), weights_only=True)

theta_low_noise, Cls_low_noise = simulations1["theta"], simulations1["x"]
theta_high_noise, Cls_high_noise = simulations2["theta"], simulations2["x"]

plt.figure(figsize=(10, 5))
for i in range(1):
    plt.plot(Cls_high_noise[i, :], marker='o', linestyle='', markersize=0.5, label="High noise")  
    plt.plot(Cls_low_noise[i, :], marker='o', linestyle='', markersize=0.5, label="Low noise")
    
plt.yscale('log')
plt.xlabel('$\ell$')
plt.ylabel('$C_\ell^{TT}$ [$\mu K^2$]')
plt.legend()
plt.show()
```
